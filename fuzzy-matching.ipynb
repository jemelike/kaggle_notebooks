```python
def generate_k_gram(value:str, k=3, ignoreCasing=True) -> [str]:
    start_index = 0
    if ignoreCasing:
        value = value.upper()
    finish_index = len(value)
    k_grams = []
    while start_index < finish_index:
        target_index = start_index + k
        k_gram = value[start_index:target_index]
        if len(k_gram) == k:
            k_grams += [k_gram]
        start_index += 1
        
    min_hash_sig = []
    hash_functions = [hashlib.sha512, hashlib.md5, hashlib.sha1, hashlib.sha256]
#     min_hash_mat = []
    for i in range(k):
        min_hash =[]
        for k_gram in k_grams:
            min_hash += [hash_functions[i](k_gram.encode()).hexdigest() ]
#         min_hash_mat +=[ k_grams]
        min_hash_sig += [min(min_hash)]
    return min_hash_sig#,min_hash_mat
```

```python
set_1 = set(generate_k_gram("James Smith", k=4))
set_2 = set(generate_k_gram("Smith James", k=4))
set_3 = set(generate_k_gram("Smith Will", k=4))
set_4 = set(generate_k_gram("Smith John", k=4))
set_5 = set(generate_k_gram("Smith Larry", k=4))
set_6 = set(generate_k_gram("Smith Jack", k=4))
```

```python
def jaccard_distance(set_1,set_2):
    common_elements = len(set_1.intersection(set_2))
    unique_elements = len(set_1.union(set_2))
    return common_elements/unique_elements
```

```python
print("James Smith", "Smith James",jaccard_distance(set_1,set_2))
print("James Smith", "Smith Will", jaccard_distance(set_1,set_3))
print("James Smith", "Smith John", jaccard_distance(set_1,set_4))
print("James Smith", "Smith Larry", jaccard_distance(set_1,set_5))
print("James Smith", "Smith Jack", jaccard_distance(set_1,set_6))
print("Smith James", "Smith Will",jaccard_distance(set_2,set_3))
print("Smith James", "Smith John",jaccard_distance(set_2,set_4))
print("Smith James", "Smith Larry",jaccard_distance(set_2,set_5))
print("Smith James", "Smith Jack",jaccard_distance(set_2,set_6))
print("Smith Will", "Smith John" , jaccard_distance(set_3,set_4))
print("Smith Will", "Smith Larry",jaccard_distance(set_3,set_5))
print("Smith Will", "Smith Jack",jaccard_distance(set_3,set_6))
print("Smith John", "Smith Larry",jaccard_distance(set_4,set_5))
print("Smith John", "Smith Jack",jaccard_distance(set_4,set_6))
print("Smith Larry","Smith Jack",jaccard_distance(set_5,set_6))
```
